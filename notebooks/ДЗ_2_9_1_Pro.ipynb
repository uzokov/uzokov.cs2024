## Задание 1

**Условие:**
Используя методы модуля ```cmath```, найти значения $\exp^{{j*2\pi k}/{n}}$ для всех целых $k$ от $1$ до $20$ и $n=4$ ($j$ - мнимая единица). Найдите количество различных элементов при условии округления действительной и мнимой частей до целого.

**Формат ответа:**
Число (например, ```2```)

**Решение:**

````python
import cmath

# Функция для вычисления значений exp(j * 2 * pi * k / n)
def calculate_values(n):
    values = set()  # Множество для хранения уникальных значений
    for k in range(1, 21):  # k от 1 до 20
        value = cmath.exp(1j * 2 * cmath.pi * k / n)
        rounded_value = (round(value.real), round(value.imag))  # Округляем действительную и мнимую части до целого
        values.add(rounded_value)
    return values

# Вычисляем значения для n=4
result = calculate_values(4)

# Выводим количество различных элементов
print(len(result))
````
